```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt
import numpy as np

data = np.load('arr_277108_R11_S11.npz')
ref = data['a']
dmag = data['b']
dmagErr = data['c']

zp, weightSum = np.average(dmag, weights=1/dmagErr**2, returned=True)
sigma = np.sqrt(1.0/weightSum)
zp2 = np.median(dmag[ref<21])
print(zp, zp2, len(dmag))
```

```python
fig, (ax0) = plt.subplots(ncols=1, figsize=(10, 10))
ax0.plot(ref, dmag, 'b+')
ax0.errorbar(ref, dmag, yerr=dmagErr, linestyle='', color='b', )
ax0.plot((14, 24), (zp, zp), 'g--', label='weighted average: %.3f'%zp)
ax0.plot((14, 24), (zp2, zp2), 'r-', label='median (mag<21): %.3f'%zp2)
ax0.set_ylim([30, 31])
ax0.tick_params(labelsize=15)
ax0.set_xlabel('Reference magnitude', fontsize=15)
ax0.set_ylabel('Reference - Source', fontsize=15)
ax0.set_title('u band 277108 R11 S11')
ax0.legend(loc='upper left', shadow=False, fontsize='x-large')
```
