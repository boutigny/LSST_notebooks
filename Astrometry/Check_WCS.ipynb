```python
import lsst.daf.persistence as dafPersist
import lsst.afw.table as afwTable
import lsst.afw.geom as afwGeom
import lsst.afw.image as afwImage
import numpy as np

visit = 11700
tract = 0
filter = 'HSC-G'

# butler = dafPersist.Butler("/sps/lsst/dev/lsstprod/hsc/SSP_UDEEP_COSMOS/output/rerun/_parent/rerun")
butler = dafPersist.Butler("/sps/lsst/dev/lsstprod/hsc/SSP_UDEEP_COSMOS/output")

for ccd in range(104) :
    if ccd == 9 :
        continue
    
    dataid = {'visit':visit, 'filter':filter, 'ccd':ccd}
    dataidNew = {'visit':visit, 'filter':filter, 'ccd':ccd, 'tract':tract}

    mdRef = butler.get("calexp_md", dataid, immediate=True)
    oldWcs = afwImage.makeWcs(mdRef)
    newWcs = butler.get('wcs', dataidNew, immediate=True).getWcs()
    
    point = afwGeom.Point2D(2000., 2000.)

    newCoord = newWcs.pixelToSky(point)
    oldCoord = oldWcs.pixelToSky(point)
    dist = newCoord.angularSeparation(oldCoord).asArcseconds()*1000.
    dRa = (newCoord.getRa()-oldCoord.getRa()).asArcseconds()*1000.
    dDec = (newCoord.getDec()-oldCoord.getDec()).asArcseconds()*1000.

    print(dist, dRa, dDec)
```
