```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt

import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.table as afwTable
import lsst.afw.image as afwImage
import math

visit = 849375
filter = 'r'

Flags = ["base_PixelFlags_flag_saturated", "base_PixelFlags_flag_cr", "base_PixelFlags_flag_interpolated",
         "base_PsfFlux_flag_edge"]

butler = dafPersist.Butler('/home/boutigny/CFHT/test_cfht/output')

Te1 = []
Te2 = []
extend = []
nSrc = 0
for ccd in range(36) :    
    vId = {'visit':visit,'ccd':ccd, 'filter':filter}
    # calexp_md : metadata
    md = butler.get("calexp_md", vId, immediate=True)
    calib = afwImage.Calib(md)
    sources = butler.get('src', vId, immediate=True)
    for src in sources :
        extend.append(src.get("base_ClassificationExtendedness_value"))
        # Select sources which look like galaxies
        if src.get("base_ClassificationExtendedness_value") < 0.5 :
            continue
        for f in Flags :
            if src.get(f) :
                continue
        # compute magnitude
        flux = src.get('base_PsfFlux_flux')
        if flux < 0. :
            print "Negative flux", ccd, flux
            continue
        mag = calib.getMagnitude(flux)
        if mag > 24 :
            continue
        nSrc += 1
        Q11 = src.get("base_SdssShape_xx")
        Q22 = src.get("base_SdssShape_yy")
        Q12 = src.get("base_SdssShape_xy")
        if Q11*Q22-Q12*Q12 < 0. :
            continue
        e1 = (Q11-Q22)/( (Q11+Q22) + 2*math.sqrt((Q11*Q22-Q12*Q12)) )
        e2 = 2*Q12/( (Q11+Q22) + 2*math.sqrt((Q11*Q22-Q12*Q12)) )
        Te1.append(e1)
        Te2.append(e2)

print "Number of sources taken into account :", nSrc
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt
import numpy as np

radius = np.sqrt(np.asarray(Te1)*np.asarray(Te1)+np.asarray(Te2)*np.asarray(Te2))

font = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'bold',
        'size'   : 25,
        }

plt.clf()
plt.rcParams['axes.linewidth'] = 2 #set the value globally

fig, (ax0) = plt.subplots(ncols=1, figsize=(10,10))
ax0.scatter(Te1, Te2, s=0.5, color='b')
ax0.set_xlim([-0.5,0.5])
ax0.set_ylim([-0.5,0.5])
#ax0.set_xlabel("mag(r)-mag(i)", fontsize=20, weight='normal')
#ax0.set_ylabel("mag(g)-mag(r)", fontsize=20, weight='normal')
#ax0.set_title("CFHTLS - D3 - mag(r) < 21", fontdict=font, y=1.02)
ax0.tick_params(labelsize=20)
plt.rcParams['axes.linewidth'] = 0.5 #set the value globally
plt.show()

print radius.ravel()
plt.clf()
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20,10))
ax1.hist(radius.ravel(), bins=100, range=[0.,1.])
ax2.hist(extend)
plt.show()
```
