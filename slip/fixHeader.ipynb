```python
import os
from astropy.io import fits

inDir = '/sps/lsst/data/nchotard/clusters/slip_data'
outDir = '/sps/lsst/data/boutigny/CFHT/slip'
badObject = {'D3 (Groth)' : 'D3_Groth'}

for inFile in os.listdir(inDir):
    if len(inFile.split('.')) == 1 or inFile.split('.')[1] != 'fits' :
        continue
    print('Checking',inFile)
    hdul = fits.open(os.path.join(inDir, inFile))
    if hdul[0].header['OBJECT'] in badObject.keys():
        for h in hdul:
            h.header['OBJECT'] = badObject[h.header['OBJECT']]
        outFile = os.path.join(outDir,inFile)
        print('Writing', outFile, hdul[0].header['OBJECT'])
        hdul.writeto(outFile, output_verify='fix+warn')
    else:
        os.symlink(os.path.join(inDir, inFile), os.path.join(outDir, inFile))                     
```
